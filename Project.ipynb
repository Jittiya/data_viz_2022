## เชื่อมตารางข้อมูล

```python
data_merged = pd.concat([data5,data2,data3,data4])   # เชื่อมข้อมูล COVID ทั้ง 4 ตารางเข้าด้วยกัน โดยใช้ pd.concat เป็นการต่อตารางในแนวนอน
data_merged
```

จากการเชื่อตารางข้อมูล COVID ทั้งหมด เราจะทำการเลือกข้อมูลที่จะนำมาวิเคราะห์ ได้แก่ ลำดับ, วันที่, เพศ, อายุ, unit(หน่วยนับอายุ), province_of_isolationและ risk

```python
data_merged1 = data_merged[['No.','announce_date','sex','age','Unit','province_of_isolation','risk']]   # ชี้ข้อมูลคอลัมน์ที่ต้องการ
data_merged1 
```

```python
data_merged1.isnull().any()  # ตรวจสอบหาค่า missing
```

จากการตรวจสอบพบว่าข้อมูลค่า missing เราจะทำการ dropna() เพื่อนำค่า missing นั้นออกไป

```python
data_merged1.dropna()  # dropna ลบค่า missing ออก
```

```python
# ตรวจสอบว่าข้อมูลหายไปเท่าไหร่
removes = data_merged1.shape[0] - data_merged1.dropna().shape[0]
print(f'size before drop = {data_merged1.shape[0]}')
print(f'size after drop = {data_merged1.dropna().shape[0]}')
print(f'we lost {100*(removes/data_merged1.shape[0])}% of data')
```

### เชื่อมตารางข้อมูล COVID กับ ข้อมูลภูมิภาค

```python
data6.isnull().any()   # ตรวจสอบหาค่า missing ของข้อมูลภูมิภาค
```

```python
data_merged2 = data_merged1.merge(data6,how='left',left_on='province_of_isolation',right_on='จังหวัด')
data_merged2
#เชื่อมตาราง data_merged1 และ data6 โดยใช้ province_of_isolation และ จังหวัด เป็นตัวเชื่อมอ้างอิง เพราะเราจะพิจารณาข้อมูลที่มีค่าข้อมูลเหมือนกัน
```

จากการเชื่อตารางข้อมูล COVID ทั้งหมด เราจะทำการเลือกข้อมูลที่จะนำมาวิเคราะห์ ได้แก่ ลำดับ, วันที่, เพศ, อายุ, unit(หน่วยนับอายุ), province_of_isolation,  riskและ ข้อมูลภูมิคภาค โดยใช้ข้อมูลแบ่งภูมิภาค 6 ภูมิภาคโดยแบ่งแบบ ภูมิศาสตร์

```python
data_merged3 = data_merged2[['No.','announce_date','sex','age','Unit','province_of_isolation','risk','จังหวัด','6 ภูมิภาค (ภูมิศาสตร์)']]
data_merged3 
```

```python
data_merged3.isnull().any()  # ตรวจสอบหาค่า missing
```

```python
data_merged3.dropna()
```

```python
# ตรวจสอบว่าข้อมูลหายไปเท่าไหร่
removes = data_merged3.shape[0] - data_merged3.dropna().shape[0]
print(f'size before drop = {data_merged3.shape[0]}')
print(f'size after drop = {data_merged3.dropna().shape[0]}')
print(f'we lost {100*(removes/data_merged3.shape[0])}% of data')
```

จากการลบค่า missing ด้วยการ dropna() พบว่าข้อมูลหายไป 12.06 %

### ลบค่า missing เฉพาะใน column ที่เราสนใจ

```python
data_merged3_nomissingAge = data_merged3.dropna(subset=['age'])
data_merged3_nomissingAge
```

```python
data_merged3_nomissingAge.isnull().any()  # ตรวจสอบหาค่า missing
```

```python
# ตรวจสอบว่าข้อมูลหายไปเท่าไหร่
removes = data_merged3_nomissingAge.shape[0] - data_merged3_nomissingAge.dropna().shape[0]
print(f'size before drop = {data_merged3_nomissingAge.shape[0]}')
print(f'size after drop = {data_merged3_nomissingAge.dropna().shape[0]}')
print(f'we lost {100*(removes/data_merged3_nomissingAge.shape[0])}% of data')
```

จากการลบค่า missing โดยลบเฉพาะคอลัมน์ที่เราสนใจพบว่าข้อมูลหาย 4.3 % แต่ยังมีข้อมูลที่มีค่า missing อยู่ จึงต้องใช้การแทนด้วย class ใหม่ (ค่าที่เหมาะสม)

### แทนค่าในคอลัมน์ที่สนใจ class ใหม่ (ค่าที่เหมาะสม)

```python
nomissingAge = data_merged3[['age']].mean() #หาค่าเฉลี่ยของอายุ
data_merged3[['age']] = data_merged3[['age']].fillna(nomissingAge) #แทนค่าเฉลี่ยของอายุลงในข้อมูลที่เป็นค่า missing
```

```python
data_merged3[['sex']] = data_merged3[['sex']].fillna('ไม่ระบุ')   #แทนค่า Missing ในคอลัมน์ sex
```

```python
data_merged3[['Unit']] = data_merged3[['Unit']].fillna('ปี')  #แทนค่า Missing ในคอลัมน์ unit
```

```python
data_merged3[['province_of_isolation']] = data_merged3[['province_of_isolation']].fillna('ไม่ระบุ')  #แทนค่า Missing ในคอลัมน์ province_of_isolation
```

```python
data_merged3[['risk']] = data_merged3[['risk']].fillna('ไม่ระบุ')  #แทนค่า Missing ในคอลัมน์ risk
```

```python
# ตรวจสอบว่าข้อมูลหายไปเท่าไหร่
removes = data_merged3.shape[0] - data_merged3.dropna().shape[0]
print(f'size before drop = {data_merged3.shape[0]}')
print(f'size after drop = {data_merged3.dropna().shape[0]}')
print(f'we lost {100*(removes/data_merged3.shape[0])}% of data')
```

จากการแทนค่า missing ในคอลัมน์ที่สนใจ แล้วทำการ dropna() ข้อมูล พบว่าข้อมูลหายไป 8.5 %
