## ต่อในแนวแกน X

```python
personal_data = data_df[['No.','sex','age','nationality']]
personal_data.head()
```

```python
covid_data = data_df[['No.','announce_date','province_of_isolation','province_of_onset','risk']].sample(frac=1)
covid_data.head()
```

### เอาตารางสองอันมาต่อกัน

```python
personalcovid_data = personal_data.merge(covid_data)     # เอาตาราง personal_data เป็นตารางหลัก แล้วเอาตาราง covid_data มาต่อ
personalcovid_data.head()
```

```python
personalcovid_data = covid_data.merge(personal_data)  # เอาตาราง covid_data เป็นตารางหลัก แล้วเอาตาราง personal_data มาต่อ
personalcovid_data.head()
```

# Save table

```python
personalcovid_data.to_csv(os.path.join(path,'my_processed_data.csv'))  # save ไปเก็บในไดร์ฟ
```

# Groupby

https://www.kaggle.com/crawford/python-groupby-tutorial#

```python
data_df.groupby('nationality').count()  # จัด group และนับค่าใน column
```

```python
data_df.groupby('nationality').mean()  # จัด group และคำนวณค่าเฉลี่ยใน column
```

```python
data_df.groupby('nationality')['age'].min()  # min ต้อง เลือก 1 column มาจัด group และอีก 1 column มาเพื่อหาค่า
```

```python
data_df.groupby('nationality').min(' ')  # จัด group และคำนวณค่าน้อยสุดใน column
```
